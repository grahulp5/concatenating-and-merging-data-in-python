```python
df_left = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                       index=['a', 'c', 'e'],
                       columns=['Ohio', 'Nevada'])
df_right1 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                         index=['b', 'c', 'd', 'e'],
                         columns=['Missouri', 'Alabama'])
df_right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [16., 17.]],
                         index=['a', 'c', 'e', 'f'],
                         columns=['New York', 'Oregon'])
display(df_left, df_right1, df_right2)
```

### Task 1 (3 marks)

Join the three dataframes using the _outer_ and _inner_ mechanisms via the __join__, __concat__, and __merge__ methods. The output from these three methods should look exactly the same.

#### Outer Join

```python
### Write your code below this comment.
# Using join
outer_join1 = df_left.join(df_right1, how='outer')

outer_join2 = outer_join1.join(df_right2, how='outer')

# Using concat
concat_outer = pd.concat([df_left, df_right1, df_right2], axis=1, join='outer')

# Using merge
outer_merge1 = pd.merge(df_left, df_right1, how='outer', left_index=True, right_index=True)

outer_merge2 = pd.merge(outer_merge1, df_right2, how='outer', left_index=True, right_index=True)



print("Outer Join:")
display(outer_join2)


print("Concat (Outer):")
display(concat_outer)


print("Merge (Outer):")
display(outer_merge2)
```

#### Inner Join

```python
### Write your code below this comment.
# Using join
inner_join1 = df_left.join(df_right1, how='inner')

inner_join2 = inner_join1.join(df_right2, how='inner')

# Using concat
concat_inner = pd.concat([df_left, df_right1, df_right2], axis=1, join='inner')

# Using merge
inner_merge1 = pd.merge(df_left, df_right1, how='inner', left_index=True, right_index=True)

inner_merge2 = pd.merge(inner_merge1, df_right2, how='inner', left_index=True, right_index=True)

# Display the results
print("Inner Join:")
display(inner_join2)


print("Concat (Inner):")
display(concat_inner)


print("Merge (Inner):")
display(inner_merge2)
```
